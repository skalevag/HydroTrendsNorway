```python
def plotKlimaservice(dfhigh,dfmed,file,ylabel,save=True,separateFigures=True,title = None,figsize=(4,3),legend=False):
    if separateFigures:
        plt.figure(figsize=figsize)
    # projections
    plt.axvline(1963,linestyle=":",color="lightgrey")
    plt.axvline(1983,linestyle=":",color="whitesmoke")
    plt.axvline(2012,linestyle=":",color="lightgrey")
    plt.fill_between(dfhigh.Year,dfhigh["10%"],dfhigh["90%"],color="darksalmon",alpha=0.1)
    plt.plot(dfhigh.Year,dfhigh["10%"],color="darksalmon",alpha=0.3)
    plt.plot(dfhigh.Year,dfhigh["90%"],color="darksalmon",alpha=0.3)
    plt.fill_between(dfmed.Year,dfmed["10%"],dfmed["90%"],color="c",alpha=0.1)
    plt.plot(dfmed.Year,dfmed["10%"],color="c",alpha=0.3)
    plt.plot(dfmed.Year,dfmed["90%"],color="c",alpha=0.3)
    plt.plot(dfhigh.Year,dfhigh.Median,label="RCP8.5",color="darksalmon")
    plt.plot(dfmed.Year,dfmed.Median,label="RCP4.5",color="c")
    plt.xlim(1900,2100)
    
    # observations
    plt.plot(dfhigh.Year,dfhigh.Obs,"k",label="Observations")
    # labels
    plt.xlabel("Year")
    plt.ylabel(ylabel)
    if legend:
        obs = mpatches.Patch(color='k', label='Observations')
        high = mpatches.Patch(color='darksalmon', label='RCP8.5')
        low = mpatches.Patch(color='c', label='RCP4.5')
        plt.legend(handles=[obs,high,low],loc="upper left")
    
    if not (title == None):
        plt.title(title,loc="left")
    
    # save plot
    if save:
        plt.savefig(f"Plots/klimaservice_ref1971-2000_{file}.png",dpi=400,bbox_inches='tight')
```

```python
plotKlimaservice(precipHigh,precipMed,"Precipitation","Precipitation anomaly (%)")
plotKlimaservice(tempHigh,tempMed,"Temperature","Temperature anomaly (°C)")
plotKlimaservice(runoffHigh,runoffMed,"Runoff","Runoff anomaly (%)",figsize=(8,3),legend=True)
```
